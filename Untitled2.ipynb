```python
import pandas as pd
import scipy.stats as stats

# Excel 파일에서 데이터 읽기
file_path = "/content/Graph.xlsx"
model = pd.read_excel(file_path)

# 온도 및 다른 변수들을 숫자(float) 형식으로 변환
x = model['Rs(J/m2)'].astype(float)
y1 = model["Penman"].astype(float)
y2 = model["FAO PMequ"].astype(float)
y3 = model["Priestley Taylor"].astype(float)
y4 = model["Hargreaves"].astype(float)
y5 = model["Fynn"].astype(float)
y6 = model["Simplified"].astype(float)

# 상관 계수 및 결정 계수 계산
correlation_coef1 = stats.pearsonr(x, y1)[0]
correlation_coef2 = stats.pearsonr(x, y2)[0]
correlation_coef3 = stats.pearsonr(x, y3)[0]
correlation_coef4 = stats.pearsonr(x, y4)[0]
correlation_coef5 = stats.pearsonr(x, y5)[0]
correlation_coef6 = stats.pearsonr(x, y6)[0]

r_squared1 = correlation_coef1 ** 2
r_squared2 = correlation_coef2 ** 2
r_squared3 = correlation_coef3 ** 2
r_squared4 = correlation_coef4 ** 2
r_squared5 = correlation_coef5 ** 2
r_squared6 = correlation_coef6 ** 2

# 결과 출력
print("Penman - 상관계수:", correlation_coef1, "결정계수:", r_squared1)
print("FAO P-M - 상관계수:", correlation_coef2, "결정계수:", r_squared2)
print("Priestley Taylor - 상관계수:", correlation_coef3, "결정계수:", r_squared3)
print("Hargreaves - 상관계수:", correlation_coef4, "결정계수:", r_squared4)
print("Fynn - 상관계수:", correlation_coef5, "결정계수:", r_squared5)
print("Simplified - 상관계수:", correlation_coef6, "결정계수:", r_squared6)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 데이터를 딕셔너리 형태로 정리
data = {
    "Model": ["FAO Panman", "FAO Penman-Monteith", "Priestley-Taylor", "Hargreaves", "Fynn", "Simplified"],
    "Rs(J/m2)": [0.976, 0.983, 0.980, 0.909, 0.974, 0.805],
    "RH(%)": [0.262, 0.172, 0.169, 0.116, 0.163, 0.082],
    "Temp('C)": [0.316, 0.347, 0.354, 0.438, 0.371, 0.598]
}

# 데이터프레임 생성
df = pd.DataFrame(data)

# PairGrid 설정
g = sns.PairGrid(df, x_vars=["Rs(J/m2)", "RH(%)", "Temp(ºC)"], y_vars=["Model"], height=4, aspect=1.2)

# Scatter plot 그리기
g.map(sns.scatterplot, s=90, alpha=0.8)

# x축 범위 설정
g.set(xlim=(0, 1))

# 각 subplot에 x축 레이블 설정
for ax, xlabel in zip(g.axes.flat, ["r_squared"] * 3):
    ax.set_xlabel(xlabel)
    ax.grid(False, axis='x')  # 각 subplot에 대해 x 축 그리드 제거

# 각 subplot에 title 설정
titles = ["Rs(J/m2)", "RH(%)", "Temp('C)"]
for ax, title in zip(g.axes.flat, titles):
    ax.set_title(title)

# 그래프 간의 간격 조정
plt.tight_layout()

# 그래프 보여주기
plt.savefig("/content/Graph.png")

plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# 데이터 준비
data = {
    'Model': ['FAO Panman', 'FAO Penman-Monteith', 'Priestley-Taylor', 'Hargreaves', 'Fynn', 'Simplified'],
    'r2(Rs-model)': [0.615, 0.616, 0.611, 0.606, 0.638, 0.542],
    'r2(RH-model)': [0.165, 0.108, 0.105, 0.077, 0.107, 0.055],
    'r2(Temp-model)': [0.220, 0.276, 0.284, 0.317, 0.255, 0.403]
}

df = pd.DataFrame(data)

# 그래프 그리기
plt.figure(figsize=(8, 7))  # 그래프 크기 설정

# 각 모델에 대해 가로 막대 그래프 그리기
sns.barplot(data=df, y='Model', x='r2(Temp-model)', color='forestgreen', alpha=0.8, label='r2(Temp-model)')
sns.barplot(data=df, y='Model', x='r2(RH-model)', color='steelblue', alpha=0.8, label='r2(RH-model)',
            left=df['r2(Temp-model)'])  # Temp-model 비율 우측에 RH-model 비율 그리기
sns.barplot(data=df, y='Model', x='r2(Rs-model)', color='firebrick', alpha=0.8, label='r2(Rs-model)',
            left=df['r2(Temp-model)'] + df['r2(RH-model)'])  # Temp-model + RH-model 비율 우측에 Rs-model 비율 그리기

plt.xlabel('Impact ratio of environmental factors on ETo')  # x축 레이블 설정
plt.ylabel('Model')  # y축 레이블 설정
plt.legend()  # 범례 표시
plt.xlim(0, 1)  # x축 범위 설정 (0부터 1까지)

# 그래프 저장 전에 여백 조정
plt.tight_layout()

plt.savefig("/content/Graph.png")
plt.show()  # 그래프 출력
```

```python
import pandas as pd
from scipy.stats import ttest_ind, mannwhitneyu

# 파일 경로
file_path = '/content/새만금 vs 전주 기상데이터.xlsx'

# 엑셀 파일 불러오기
df = pd.read_excel(file_path)

# 비교할 열 목록
columns_to_compare = {
    'Solar Radiation (W/m2)': ['S Solar Radiation (W/m2)', 'J Solar Radiation (W/m2)'],
    'Wind Speed (m/s)': ['S Wind Speed (m/s)', 'J Wind Speed (m/s)'],
    'Air Temperature (°C)': ['S Air Temperature (°C)', 'J Air Temperature (°C)'],
    'Atmospheric Pressure (kPa)': ['S Atmospheric Pressure (kPa)', 'J Atmospheric Pressure (kPa)'],
    'Relative Humidity (%)': ['S RH(%)', 'J RH(%)']
}

# 각 변수에 대해 t-test와 Mann-Whitney U test 수행
for variable, columns in columns_to_compare.items():
    s_data = df[columns[0]].dropna()
    j_data = df[columns[1]].dropna()

    # t-test 수행
    t_stat, p_value = ttest_ind(s_data, j_data)

    # Mann-Whitney U 검정 수행
    u_stat, p_value_u = mannwhitneyu(s_data, j_data)

    # 결과 출력 (소수점 5자리로 포맷팅)
    print(f"\n{variable} comparison:")
    print("T-test results:")
    print("t-statistic:", round(t_stat, 5))
    print("p-value:", round(p_value, 5))

    print("\nMann-Whitney U test results:")
    print("U statistic:", round(u_stat, 5))
    print("p-value:", round(p_value_u, 5))
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.stats import pearsonr

# 파일 경로
file_path = '/content/새만금 ETo(model).xlsx'

# 엑셀 파일 불러오기
df = pd.read_excel(file_path)

# NaN 값을 각 열의 평균값으로 대체
df = df.fillna(df.mean())

# Lysimeter ETo와 각 모델의 RMSE, MAPE 및 피어슨 상관계수 계산
models = ['FAO Penman', 'Penman', 'FAO P-M', 'Simplified', 'Fynn', 'K-P', 'Radiation',
          'D-K', 'Makkink', 'P-T', 'Turc', 'H-S', 'Hargreaves', 'McCloud']

metrics = {'Model': [], 'RMSE': [], 'MAPE': [], 'Pearson': []}
lysimeter_values = df['Lysimeter ETo']

for model in models:
    predictions = df[model]
    rmse = np.sqrt(mean_squared_error(lysimeter_values, predictions))
    mape = np.mean(np.abs((lysimeter_values - predictions) / lysimeter_values)) * 100
    pearson_corr, _ = pearsonr(lysimeter_values, predictions)

    metrics['Model'].append(model)
    metrics['RMSE'].append(rmse)
    metrics['MAPE'].append(mape)
    metrics['Pearson'].append(pearson_corr)

# 데이터프레임으로 변환
metrics_df = pd.DataFrame(metrics).sort_values(by='RMSE')

# 스타일 설정
sns.set_theme(style="whitegrid")
plt.figure(figsize=(30, 8))

# RMSE 시각화
plt.subplot(1, 3, 1)
ax1 = sns.barplot(x='RMSE', y='Model', data=metrics_df, palette='Blues', edgecolor='black')
plt.xlim(0, metrics_df['RMSE'].max() + 0.5)
plt.xlabel('RMSE (mm)', fontsize=16, labelpad=10)
plt.ylabel('Model', fontsize=16)
plt.axvline(x=metrics_df['RMSE'].min(), color='red', linestyle='--', linewidth=2, label='Best RMSE')
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.5)

# RMSE 값 표시
for p in ax1.patches:
    ax1.annotate(f"{p.get_width():.2f} mm", (p.get_width() + 0.05, p.get_y() + p.get_height() / 2),
                  ha='left', va='center', color='black', fontsize=12)

# 그래프 테두리 두껍게 설정
for spine in ax1.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1)

# MAPE 시각화
plt.subplot(1, 3, 2)
ax2 = sns.barplot(x='MAPE', y='Model', data=metrics_df, palette='Blues', edgecolor='black')
plt.xlim(0, metrics_df['MAPE'].max() + 10)
plt.xlabel('MAPE (%)', fontsize=16, labelpad=10)
plt.ylabel('Model', fontsize=16)
plt.axvline(x=metrics_df['MAPE'].min(), color='red', linestyle='--', linewidth=2, label='Best MAPE')
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.5)

# MAPE 값 표시
for p in ax2.patches:
    ax2.annotate(f"{p.get_width():.2f} %", (p.get_width() + 1, p.get_y() + p.get_height() / 2),
                  ha='left', va='center', color='black', fontsize=12)

# 그래프 테두리 두껍게 설정
for spine in ax2.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1)

# 피어슨 상관계수 시각화
plt.subplot(1, 3, 3)
ax3 = sns.barplot(x='Pearson', y='Model', data=metrics_df, palette='Blues', edgecolor='black')
plt.xlim(-1, 1)  # 피어슨 상관계수는 -1에서 1 사이
plt.xlabel('Pearson Correlation', fontsize=16, labelpad=10)
plt.ylabel('Model', fontsize=16)
plt.axvline(x=metrics_df['Pearson'].max(), color='red', linestyle='--', linewidth=2, label='Best Pearson')
plt.legend()
plt.grid(axis='x', linestyle='--', alpha=0.5)

# 피어슨 상관계수 값 표시
for p in ax3.patches:
    ax3.annotate(f"{p.get_width():.2f}", (p.get_width() + 0.02, p.get_y() + p.get_height() / 2),
                  ha='left', va='center', color='black', fontsize=12)

# 그래프 테두리 두껍게 설정
for spine in ax3.spines.values():
    spine.set_edgecolor('black')
    spine.set_linewidth(1)

# 그래프 저장
plt.savefig('Saemangeum_vs_model(RMSE_MAPE_Pearson).png')

plt.tight_layout()
plt.show()
```
